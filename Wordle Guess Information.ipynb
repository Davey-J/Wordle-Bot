```python
import pandas as pd
import numpy as np
import string
import itertools
import time
```

```python
raw_wordle_words = pd.read_csv('Word Data/wordle_words.csv')
wordle_words_counted = pd.read_csv('Word Data/counted_wordle_words.csv')
wordle_words_counted = wordle_words_counted.drop(columns=['Unnamed: 0','count'])
wordle_words_counted
```

```python
# Probability of result = number of count/total words
# Information of result = -log(number of count/total words)
# Entropy of word = -sum(prob * information)
```

```python
total_count = wordle_words_counted.drop(columns=['word']).sum(axis=1)
total_count = total_count[0]
total_count
```

```python
def calculate_entropy(word):
    counts = wordle_words_counted[wordle_words_counted["word"]==word].drop(columns=["word"])
    entropy = counts.apply(lambda x: x/total_count * -np.log2(x/total_count), axis=1).sum(axis=1)
    return entropy

calculate_entropy("about")
```

```python
wordle_words_counted["entropy"] = wordle_words_counted.drop(columns=["word"]).apply(lambda x: x/total_count * -np.log2(x/total_count), axis=1).sum(axis=1)
wordle_words_counted
```

```python
wordle_words_counted.to_csv('Word Data/wordle_words_entropy.csv')
```
