```python
import pandas as pd
import numpy as np
import string
import itertools
import time
```

```python
raw_wordle_words = pd.read_csv('Word Data/wordle_words.csv')
wordle_words = raw_wordle_words.copy()
wordle_words
```

```python
letters = ["G", "Y", "N"]
possible_results = itertools.product(letters, repeat=5)
possible_results = [''.join(seq) for seq in possible_results]
possible_results
```

```python
def check_result_frequency(word, result):
    possibles = raw_wordle_words
    if result == 'GGGGG':
        return 1
    for letter in range(5):
        if result[letter] == "G":
            possibles = possibles[possibles["word"].str[letter] == word[letter]]
        elif result[letter] == "Y":
            possibles = possibles[possibles["word"].str.contains(word[letter])]
            possibles = possibles[possibles["word"].str[letter] != word[letter]]
        elif result[letter] == "N":
            possibles = possibles[~possibles["word"].str.contains(word[letter])]
    return len(possibles)
```

```python
test_def = raw_wordle_words[raw_wordle_words["word"].str.contains("x")]
test_def
```

```python
pd.options.mode.copy_on_write = True
```

```python
test_def["GNNNY"] = test_def["word"].apply(lambda w: check_result_frequency(w, "GNNNY"))
test_def
```

```python
possible_result = "YGNNY"
wordle_words[f"{possible_result} Count"] = wordle_words["word"].apply(lambda w: check_result_frequency(w, possible_result))
wordle_words
```

```python
possible_results_green_first = [x for x in possible_results if x[0] == "G"]
possible_results_yellow_first = [x for x in possible_results if x[0] == "Y"]
possible_results_grey_first = [x for x in possible_results if x[0] == "N"]
possible_results_grey_first
```

```python
start_time_green = time.time()
loop_times_green = [start_time_green]
for possible_result in possible_results_green_first:
    loop_start_time = time.time()
    wordle_words[f"{possible_result} Count"] = wordle_words["word"].apply(lambda w: check_result_frequency(w, possible_result))
    loop_times_green.append(time.time())
    print("{0} Done in {1} seocnds".format(possible_result, time.time() - loop_start_time))
wordle_words
```

```python
new_wordle_words = wordle_words.copy()
new_wordle_words
```

```python
start_time_yellow = time.time()
loop_times_yellow = [start_time_yellow]
for possible_result in possible_results_yellow_first:
    loop_start_time = time.time()
    new_wordle_words[f"{possible_result} Count"] = new_wordle_words["word"].apply(lambda w: check_result_frequency(w, possible_result))
    loop_times_yellow.append(time.time())
    print("{0} Done in {1} seocnds".format(possible_result, time.time() - loop_start_time))
new_wordle_words
```

```python
new_new_wordle_words = new_wordle_words.copy()
new_new_wordle_words
```

```python
start_time_grey = time.time()
loop_times_grey = [start_time_grey]
for possible_result in possible_results_grey_first:
    loop_start_time = time.time()
    new_new_wordle_words[f"{possible_result} Count"] = new_new_wordle_words["word"].apply(lambda w: check_result_frequency(w, possible_result))
    loop_times_grey.append(time.time())
    print("{0} Done in {1} seocnds".format(possible_result, time.time() - loop_start_time))
new_new_wordle_words
```
